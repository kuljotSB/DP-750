### Load CSV (Volume) -> Bronze Tables

```python
def load_csv(path, table):
    (spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(path)
        .write.mode("overwrite")
        .saveAsTable(table))

base = "/Volumes/retail_job_lab/bronze/raw_files"

load_csv(f"{base}/transactions.csv", "retail_job_lab.bronze.transactions")
load_csv(f"{base}/stores.csv", "retail_job_lab.bronze.stores")
load_csv(f"{base}/products.csv", "retail_job_lab.bronze.products")

print("Bronze tables created")
```
