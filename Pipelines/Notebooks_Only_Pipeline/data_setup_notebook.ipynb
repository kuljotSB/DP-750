## Notebook 1 - Load Data and Create UC Structure

### Setup Catalog and Volume 

```python
# Create catalog + schemas

spark.sql("""
CREATE SCHEMA IF NOT EXISTS retail_job_lab.bronze;
""")

spark.sql("""
CREATE SCHEMA IF NOT EXISTS retail_job_lab.silver;
""")

spark.sql("""
CREATE SCHEMA IF NOT EXISTS retail_job_lab.gold;
""")

spark.sql("""
CREATE VOLUME IF NOT EXISTS retail_job_lab.bronze.raw_files;
""")

print("Catalog and schemas created")
```

### Load Files from GitHub to UC Volume

```python
import requests

# Define the current catalog
catalog_name = "retail_job_lab"

# Define the base path using the current catalog
volume_base = f"/Volumes/{catalog_name}/bronze/raw_files"

# List of files to download
files = ["products.csv", "stores.csv", "transactions.csv"]

# Download each file
for file in files:
    url = f"https://raw.githubusercontent.com/kuljotSB/DP-750/refs/heads/main/Pipelines/Data/{file}"
    response = requests.get(url)
    response.raise_for_status()

    # Write to Unity Catalog volume
    with open(f"{volume_base}/{file}", "wb") as f:
        f.write(response.content)
```
