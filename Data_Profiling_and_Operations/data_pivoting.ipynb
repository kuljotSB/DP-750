## `Pivot` and `Unpivot` Data

### Pivot orders by cuisine

```python
%sql

SELECT *
FROM (
  SELECT city, cuisine
  FROM food_delivery_lab.bronze.orders
  WHERE delivery_status = 'Delivered'
)
PIVOT (
  COUNT(*) FOR cuisine IN ('American','Mexican','Italian','BBQ','Asian')
);
```

### Unpivot Payments

```python
%sql

SELECT *
FROM food_delivery_lab.bronze.payments
UNPIVOT (
  value FOR metric IN (amount_usd)
);
```
