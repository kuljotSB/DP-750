## View-Level Security

### Creating a View from Table with Filtering and Masking Policies

```python
%sql 
CREATE OR REPLACE VIEW default.sales_raw_view AS 
SELECT 
    SalesOrderNumber,
    CustomerName,
    CASE 
        WHEN current_user() = 'kuljot.bakshi@outlook.com' THEN '****@***.****'
        ELSE Email
    END AS Email,
    Item
FROM default.sales_raw
WHERE 
    CASE WHEN 
        current_user() = 'admin@gmail.com' THEN TRUE
        ELSE CustomerName = 'Mariah Foster'
    END;
```

### Query the View to see the applied policies in action

```python
%sql
SELECT * FROM default.sales_raw;
```
