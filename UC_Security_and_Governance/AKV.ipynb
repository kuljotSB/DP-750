## Azure Key Vault-backed Secret Scope in Azure Databricks

### Listing all the Secret Scopes in the Databricks Workspace

```python
scope = dbutils.secrets.listScopes()[0].name
print(scope)
```

```python
dbutils.secrets.list(scope)
```

### Accessing Secrets from Azure Key Vault-backed Secret Scope in Azure Databricks

```python
username = dbutils.secrets.get(scope=scope, key="username")
password = dbutils.secrets.get(scope=scope, key="password")

print("username: ", username)
print("password: ", password)
```

```python
%sql
SELECT secret('demo-kv', 'username') AS db_username;
```
